```python
from core.constants import EXCHANGES
from pprint import pprint
import pandas as pd
```

```python
pprint(EXCHANGES['KUCOIN'])
```

```python
df = pd.DataFrame(columns=['EXCHANGE','requiredCredentials', 'rateLimit', 'has', 'margin', 'fetchBorrowRate'])

for EXCHANGE in EXCHANGES:

    if EXCHANGES[EXCHANGE]['requiredCredentials']:
        _requiredCredentials = True
    else:
        _requiredCredentials = False

    if EXCHANGES[EXCHANGE]['rateLimit']:
        _rateLimit = True
    else:
        _rateLimit = False

    if EXCHANGES[EXCHANGE]['has']:
        _has = True

        if 'margin' in EXCHANGES[EXCHANGE]['has']:
            _margin = EXCHANGES[EXCHANGE]['has']['margin']
        else:
            _margin = None

        if 'fetchBorrowRate' in EXCHANGES[EXCHANGE]['has']:
            _fetchBorrowRate = EXCHANGES[EXCHANGE]['has']['fetchBorrowRate']
        else:
            _fetchBorrowRate = None
    else:
        _has = False

    df.loc[len(df.index)] = [EXCHANGE, _requiredCredentials, _rateLimit, _has, _margin, _fetchBorrowRate]


    print("----------------------------------------\n"
          f"{EXCHANGE}:\n"
          f"  requiredCredentials: {_requiredCredentials}\n"
          f"            rateLimit: {_rateLimit}\n"
          f"                  has: {_has}\n"
          f"               margin: {_margin}\n"
          f"      fetchBorrowRate: {_fetchBorrowRate}")


```
